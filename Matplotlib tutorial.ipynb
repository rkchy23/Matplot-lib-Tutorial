```python
x=np.linspace(0,10,50)
```

```python
x
```

```python
y=np.sin(x)
z=np.cos(x)
a=np.tan(x)
```

```python
plt.plot(x,y,label="sinx")
plt.plot(x,z,label="cosx")
plt.xlabel("Radian-axis")
plt.ylabel("y-axis")
plt.title("Sin graph")
plt.legend(loc="center right",ncol=1)
plt.grid(b=True,color="red")
```

```python
plt.plot(x,a)
plt.minorticks_on()
plt.grid(b=True)
plt.grid(b=True,which="minor",color="red",)
plt.grid(b=True,color="blue")
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("Radian-axis")
plt.xlim(0,10)
plt.ylabel("y-axis")
plt.title("Sin graph")
plt.grid(b=True,color="red",axis="y")
```

```python
x=np.random.randn(1000)
y=np.sin(x)
z=np.cos(x)
plt.hist(x,bins="auto",ec="k")
plt.hist(y,bins="auto",ec="k")
plt.hist(z,bins="auto",ec="k")

plt.xlabel("age")
plt.ylabel("count of people")
plt.title("Population Count")

```
